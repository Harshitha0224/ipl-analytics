```python
import pandas as pd
import numpy as np

matches = pd.read_csv('../data/raw/matches.csv')
deliveries = pd.read_csv('../data/raw/deliveries.csv')
```

```python
matches['city']             = matches['city'].fillna('Unknown')
matches['player_of_match']  = matches['player_of_match'].fillna('Unknown')
matches['winner']           = matches['winner'].fillna('No Result')
matches['result_margin']    = matches['result_margin'].fillna(0)
matches['method']           = matches['method'].fillna('Normal')

print("✅ Missing values fixed!")
print("\nMissing values remaining:")
print(matches.isnull().sum())
```

```python
team_name_map = {
    'Delhi Daredevils'          : 'Delhi Capitals',
    'Deccan Chargers'           : 'Sunrisers Hyderabad',
    'Pune Warriors'             : 'Rising Pune Supergiants',
    'Rising Pune Supergiant'    : 'Rising Pune Supergiants',
    'Kings XI Punjab'           : 'Punjab Kings',
    'Royal Challengers Bangalore' : 'Royal Challengers Bengaluru',
    'Gujarat Lions'               : 'Gujarat Titans'
}

matches['team1']       = matches['team1'].replace(team_name_map)
matches['team2']       = matches['team2'].replace(team_name_map)
matches['winner']      = matches['winner'].replace(team_name_map)
matches['toss_winner'] = matches['toss_winner'].replace(team_name_map)

print("✅ Team names standardized!")
print(f"\nTotal unique teams: {matches['team1'].nunique()}")
print("\nAll unique teams:")
print(sorted(matches['team1'].unique()))
```

```python
matches['date'] = pd.to_datetime(matches['date'])

matches['year'] = matches['date'].dt.year
```

```python
deliveries.head()
```

```python
if 'batter' in deliveries.columns:
    deliveries.rename(columns={'batter': 'batsman'}, inplace=True)
if 'batter' in matches.columns:
    matches.rename(columns={'batter': 'batsman'}, inplace=True)
```

```python
print("=== FINAL MATCHES INFO ===")
print(f"Shape: {matches.shape}")
print(f"Missing values: {matches.isnull().sum().sum()}")
print(f"Years covered: {matches['year'].min()} to {matches['year'].max()}")
print(f"Total teams: {matches['team1'].nunique()}")

print("\n=== FINAL DELIVERIES INFO ===")
print(f"Shape: {deliveries.shape}")
print(f"Total runs in dataset: {deliveries['total_runs'].sum()}")
```

```python
matches['target_runs']  = matches['target_runs'].fillna(0)
matches['target_overs'] = matches['target_overs'].fillna(0)
total_missing = matches.isnull().sum().sum()
print(f"Total missing values remaining: {total_missing}")
```

```python
matches.to_csv('../data/processed/matches_clean.csv', index=False)
deliveries.to_csv('../data/processed/deliveries_clean.csv', index=False)
```
