```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style='darkgrid')

plt.rcParams['figure.figsize'] = (12, 5)
```

```python
matches    = pd.read_csv('../data/processed/matches_clean.csv')
deliveries = pd.read_csv('../data/processed/deliveries_clean.csv')
```

```python
# WHAT: Horizontal bar chart showing which team won most matches
# WHY:  Bar charts are best for COMPARING categories (teams in this case)
# HOW:  value_counts() gives us wins per team, then we plot it

win_counts = matches['winner'].value_counts().head(10)

# Create the figure and axes
# fig  → the entire canvas
# ax   → the actual chart area inside the canvas
fig, ax = plt.subplots(figsize=(12, 6))

# sns.barplot → seaborn's bar chart
# x           → values (number of wins) go on x axis
# y           → categories (team names) go on y axis
# palette     → color scheme ('viridis' is a popular professional palette)
sns.barplot(
    x=win_counts.values,
    y=win_counts.index,
    hue=win_counts.index,      
    palette='viridis',
    legend=False,
    ax=ax
)

# Add labels and title to make chart readable
ax.set_title('🏆 Most IPL Wins by Team (All Time)', fontsize=16, fontweight='bold')
ax.set_xlabel('Number of Wins', fontsize=12)
ax.set_ylabel('Team', fontsize=12)

# Add value labels at the end of each bar so reader doesnt have to guess
for i, v in enumerate(win_counts.values):
    ax.text(v + 1, i, str(v), va='center', fontweight='bold')

plt.tight_layout()  # prevents labels from getting cut off
plt.savefig('../data/processed/chart1_wins.png', dpi=150)  # save chart as image
plt.show()
```

```python
toss_pref = matches['toss_decision'].value_counts()

fig, ax = plt.subplots(figsize=(8, 8))

ax.pie(toss_pref.values,
       labels=toss_pref.index,
       autopct='%1.1f%%',
       colors=['#2196F3', '#FF9800'],
       startangle=90,
       explode=(0.05, 0.05))

ax.set_title('🎯 Toss Decision — Field vs Bat', fontsize=16, fontweight='bold')

plt.savefig('../data/processed/chart2_toss.png', dpi=150)
plt.show()
```

```python
top_batsmen = (deliveries.groupby('batsman')['batsman_runs']
                          .sum()
                          .sort_values(ascending=False)
                          .head(10)
                          .reset_index())
top_batsmen.columns = ['Batsman', 'Total Runs']

fig, ax = plt.subplots(figsize=(12, 6))

sns.barplot(data=top_batsmen,
            x='Total Runs',
            y='Batsman',
            palette='rocket',
            ax=ax)

ax.set_title('🏏 Top 10 IPL Run Scorers (All Time)', fontsize=16, fontweight='bold')
ax.set_xlabel('Total Runs', fontsize=12)
ax.set_ylabel('Batsman', fontsize=12)

# Add run values on bars
for i, v in enumerate(top_batsmen['Total Runs']):
    ax.text(v + 50, i, str(v), va='center', fontweight='bold')

plt.tight_layout()
plt.savefig('../data/processed/chart3_batsmen.png', dpi=150)
plt.show()
```

```python
matches['date'] = pd.to_datetime(matches['date'])
matches['year'] = matches['date'].dt.year

merged = deliveries.merge(matches[['id', 'year']], 
                          left_on='match_id',   # column in deliveries
                          right_on='id')         # column in matches

runs_per_year = (merged.groupby('year')['total_runs']
                        .sum()
                        .reset_index())

fig, ax = plt.subplots(figsize=(12, 5))

# sns.lineplot → line chart
# marker='o'   → adds a dot at each data point
sns.lineplot(data=runs_per_year,
             x='year',
             y='total_runs',
             marker='o',
             color='#E91E63',
             linewidth=2.5,
             ax=ax)

ax.set_title('📈 Total Runs Scored Per IPL Season', fontsize=16, fontweight='bold')
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Total Runs', fontsize=12)

# Add value labels on each point
for _, row in runs_per_year.iterrows():
    ax.text(row['year'], row['total_runs'] + 500, 
            str(int(row['total_runs'])), 
            ha='center', fontsize=8)

plt.tight_layout()
plt.savefig('../data/processed/chart4_runs_season.png', dpi=150)
plt.show()
```
